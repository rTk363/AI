```python
data=pd.read_csv("TITANIC.csv")
```

# Problem Statement

```python
print("Problem Statement")
print("The survived passengers are: ")
data[data['Survived'] == 1]['Name']
```

## Part I

```python
print("The no of children who didnt survive")
data[(data['Survived']==0) & (data['Age']<18)][['Name','Age']]
```

```python
print("The no of women who didnt survive")
data[(data['Survived']==0) & (data['Sex']=='female') & (data['Age']>=18)][['Name','Age','Sex']]
```

```python
print("The no of Upper class people who didnt survive")
data[(data['Survived']==0) & (data['Pclass']==3)][['Name','Pclass','Sex']]
```
